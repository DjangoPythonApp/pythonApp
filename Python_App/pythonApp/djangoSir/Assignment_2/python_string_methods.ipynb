
# Python String Methods

---

# 1. What does the `strip()` method do?

The `strip()` method removes extra spaces (or specified characters) from both the beginning and end of a string.

### Example

```python
text = "   Hello World   "

print(text.strip())
```

### Output

```python
Hello World
```

It does **not** remove spaces from the middle.

### Another Example

```python
text = "###Python###"

print(text.strip("#"))
```

### Output

```python
Python
```

---

# 2. What is the difference between `lstrip()` and `rstrip()`?

| Method     | Meaning                                    |
| ---------- | ------------------------------------------ |
| `lstrip()` | Removes characters from the **left side**  |
| `rstrip()` | Removes characters from the **right side** |

### Example

```python
text = "   Python   "

print(text.lstrip())   # Left strip
print(text.rstrip())   # Right strip
```

### Output

```python
Python   
   Python
```

# 3. What does the `replace()` method do?

The `replace()` method replaces one part of a string with another.

### Syntax

```python
string.replace(old, new)
```

### Example

```python
text = "I like Java"

print(text.replace("Java", "Python"))
```

### Output

```python
I like Python
```

### Replace Multiple Occurrences

```python
text = "apple apple apple"

print(text.replace("apple", "mango"))
```

### Output

```python
mango mango mango
```

# 4. What is the use of the `find()` method?

The `find()` method searches for a substring and returns its position (index).

### Example

```python
text = "Hello Python"

print(text.find("Python"))
```

### Output

```python
6
```

Because `"Python"` starts at index 6.

### If Not Found

```python
print(text.find("Java"))
```

### Output

```python
-1
```

---

# 5. What is the difference between `find()` and `index()`?

| Feature            | `find()`     | `index()`   |
| ------------------ | ------------ | ----------- |
| Returns position   | Yes          | Yes         |
| If value not found | Returns `-1` | Gives error |

### Example

```python
text = "Python"

print(text.find("Java"))
```

### Output

```python
-1
```

### Using `index()`

```python
print(text.index("Java"))
```

### Output

```python
ValueError
```

So:

* Use `find()` when you want safe searching.
* Use `index()` when the value must exist.

# 6. What does `startswith()` check?

The `startswith()` method checks whether a string starts with a specific value.

### Example

```python
text = "Python Programming"

print(text.startswith("Python"))
```

### Output

```python
True
```

### Another Example

```python
print(text.startswith("Java"))
```

### Output

```python
False
```

---

# 7. What does `endswith()` check?

The `endswith()` method checks whether a string ends with a specific value.

### Example

```python
text = "notes.pdf"

print(text.endswith(".pdf"))
```

### Output

```python
True
```

### Another Example

```python
print(text.endswith(".txt"))
```

### Output

```python
False
```
