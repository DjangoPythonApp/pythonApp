# 20 Practice Problems on Comprehension in Python

## 1. Square Numbers

```python
number:list = [1,2,3,4,5]
Square : list = [n*n for n in number]
print(Square)
```

## 2.Square of even numbers

```python
number:list = [1,2,3,4,5]
Square : list = [n*n for n in number if n%2 == 0]
print(Square)
```

## 3. Square of  Odd numbers

```python
number:list = [1,2,3,4,5]
Square : list = [n*n for n in number if n%2 != 0]
print(Square)
```

## 4. Convert Celsius to Fahrenheit

Convert all temperatures into Fahrenheit using comprehension.

Formula: f=(5/9)*c+32

celsius = [0, 10, 20, 30, 40]


```python
celsius : list  = [0.10,20,30,40]

fahrenheit : list = [  (5/9)*i+32 for i in celsius ]
print(fahrenheit)
```

## 5. Positive and Negative

Given:

numbers = [10, -5, 0, 7, -2]

Create a new list:

"Positive" if number > 0
"Negative" if number < 0
"Zero" otherwise

```python
numbers : list = [10, -5, 0, 7, -2]
PN = ["Positive" if n > 0 else "Negative" if n < 0 else "Zero" for n in numbers]
print(PN)
```

## 6. Multiplication Table

Create a list containing multiplication table of 5 from 1 to 10.

Expected Output

[5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

```python
table_of_5 : list = [  5*i for i in range(1,10+1)]
print(table_of_5)
```

# 7. Extract Vowels

Given a string:

word = "Programming"

Extract only vowels using list comprehension.

```python
word:str = 'Programming'
vowel:list = [ i  for i in word if i.upper() == "A" or i.upper() == "E" or i.upper()== "I" or  i.upper() == "O" or i.upper() == "U" ] 
print(vowel)
```

# 8. Remove Spaces

Given:

sentence = "Python is awesome"

Create a list of characters excluding spaces.

```python
sentence :str = "Python is awesome"

removeSpace: list  = [  i for i in sentence if i != " "]
print(removeSpace)
```

# 9. Flatten Nested List

Given:

matrix = [[1,2], [3,4], [5,6]]

Convert it into:

[1,2,3,4,5,6]

```python
matrix:list  = [[1,2], [3,4], [5,6]]

flatten_List : list = [j for i in matrix for j in i ]
print(flatten_List)
```

# 10. Matrix Transpose

Transpose this matrix using nested list comprehension:

## matrix = [
    [1,2,3],
     [4,5,6]
## ]

## Normal form

```python
matrix = [
    [1,2,3],
    [4,5,6]
]
new_matrix : list = []
new_sub_1 : list = []
new_sub_2 : list = []
new_sub_3 : list = []

for i in range(len(matrix)+1):
   

    if i == 0:
       for j in range(len(matrix)):
       
          new_sub_1.append(matrix[j][i])
       new_matrix.append(new_sub_1)
       

    if i == 1:
       for j in range(len(matrix)):
       
          new_sub_2.append(matrix[j][i])
       new_matrix.append(new_sub_2)
       

    if i == 2:
       for j in range(len(matrix)):
       
          new_sub_3.append(matrix[j][i])
       new_matrix.append(new_sub_3)
       

print(new_matrix)
```
